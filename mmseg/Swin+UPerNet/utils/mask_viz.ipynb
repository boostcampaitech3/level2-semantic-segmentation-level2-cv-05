### 라이브러리 

```python
import cv2
import random
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib import gridspec
```

### image와 mask 시각화

```python


index = random.randint(0, 2616) # train dataset 중 랜덤하게 index 추출 

fig = plt.figure(figsize=(16, 16)) 
gs = gridspec.GridSpec(nrows=1, # row 몇 개 
                       ncols=2, # col 몇 개 
                       width_ratios=[10, 10, 1]
                      )

ax = []
for i in range(2):
    ax.append(plt.subplot(gs[i]))

img = mpimg.imread(f'/opt/ml/input/copy_paste/images/copy_paste_{index:04}.jpg') # image 경로 
ax[0].imshow(img)

mask = mpimg.imread(f'/opt/ml/input/copy_paste/annotations/copy_paste_{index:04}.png') # mask 경로 
ax[1].imshow(mask)

plt.show()
```
